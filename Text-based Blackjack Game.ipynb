```python
import random
```

```python
#card class
class Card():
    def __init__(self,suit,num):
        self.suit = suit
        self.num = num
        
        self.upper_card = {11:'Jack',12:'Queen',13:'King',14:'Ace'}
        
        #checking for jack, queen king
        if num in [11,12,13]:
            self.game_val = 10
        elif num ==14:
            self.game_val = 11
            #could also be 1 but will evaluate that later in the hand class
        else:
            self.game_val = num
            
    def show(self):
        if self.num in self.upper_card:
            self.cdshow = self.upper_card[self.num]
        else:
            self.cdshow = self.num
        print(f"{self.cdshow} of {self.suit}")
        
#deck
class Deck():
    def __init__(self):
        self.cards = []
        self.build()
        
    def build(self):
        for s in ["Spades","Clubs","Diamonds","Hearts"]:
            for v in range(1,15):
                add_card = Card(s,v)
                self.cards.append(add_card)
                
    def show(self):
        for c in self.cards:
            c.show()
            
    def shuffle(self):
        for i in range(len(self.cards)-1,0,-1):
            r = random.randint(0,i)
            self.cards[i],self.cards[r] = self.cards[r],self.cards[i]
            
    def draw(self):
        return self.cards.pop()
    
class Player():
    
    def __init__(self,account=500):
        self.account = account 
        self.bet_amount = 0
        
    def bet(self):
        self.bet_amount = int(input('How much do you want to bet?:'))
        
    def lose_bet(self):
        self.account = self.account - self.bet_amount
        
    def win_bet(self):
        self.account = self.account + self.bet_amount
        
    def show_act(self):
        print(self.account)  
        
class Hand():
    
    def __init__(self,deck):
        self.hand = []
        self.deck = deck
        
        #want to initiate with two cards in the beginning
        self.draw(self.deck)
        self.draw(self.deck)
        
    def draw(self, deck):
        self.hand.append(deck.draw())
        self.calculate()

    def show(self):
        self.calculate()
        for c in self.hand:
            c.show()
        print(f"total value of cards: {self.tval}")
        
    def calculate(self):
        self.tval = 0
        for c in self.hand:
            self.tval += c.game_val
        for c in self.hand:
            if c.num==14 and self.tval>21:
                c.game_val = 1 
                self.tval = self.tval - 10    
            
class PlayerHand(Hand, Player):
    
    def __init__(self, deck):
        Hand.__init__(self,deck)
        
        self.busted=False
        
    def calculate(self):
        super().calculate()
        self.self_eval()
    
    def self_eval(self):
        if self.tval >21:
            self.busted = True    

            
class Dealer(Hand):
    
    def __init__(self,deck,player_hand):
        Hand.__init__(self,deck)
        self.player_hand = player_hand
        self.player_hand.calculate()
        self.p_total_value = player_hand.tval
        print('made it past initialization')
    
        while self.p_total_value > self.tval:
            self.draw(deck)
            print('making it')
                

def play(player1):
    deck=Deck()
    deck.shuffle()
    player_hand=PlayerHand(deck)
    player1.bet()
    player_hand.show()
    
    while player_hand.busted ==False:
        hit_or_stay = input('Would you like to hit or stay? (h/s):')
        while hit_or_stay not in ['h','s']:
            hit_or_stay = input('Would you like to hit or stay? (h/s):')
        if hit_or_stay =='h':
            player_hand.draw(deck)
            player_hand.show()
        elif hit_or_stay =='s':
            dealer=Dealer(deck,player_hand)

            #evaluate
            if dealer.p_total_value > dealer.tval:
                print(f"Player wins against Dealer's {dealer.tval}")
                player1.win_bet()
                player1.show_act()
                break
            elif dealer.p_total_value < dealer.tval and dealer.tval <=21:
                print(f"Dealer wins with {dealer.tval}")
                player1.lose_bet()
                player1.show_act()
                break
            elif dealer.tval >21:
                print("Dealer Busts! Player wins") 
                player1.win_bet()
                player1.show_act()
                break
            else:
                print('It was a draw!')
                player1.show_act()
                break         
    else:
        #means they busted
        print("You busted! Game over!")
        player1.lose_bet()
        player1.show_act()
                
            
if __name__=="__main__":
    player = Player()
    play(player)
    keep_playing = input("Would you like to keep playing?(y/n):")
    while keep_playing =='y':
        play(player)
        keep_playing = input("Would you like to keep playing?(y/n):")
            
    else:
        print(f'Thanks for playing! You finished with {player.account}')
```
